```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

lam = np.linspace(200,800,301)
Tper_L = ['1', '2', '3', '4', '5', '6']
Tper_T = ['1', '2', '3', '4', '5', '6']
for i in {1, 2, 3, 4, 5, 6}:
    Tper_L[i-1] = []
    #print(Tper_L[1])
    Tper_T[i-1] = []
    with open('B'+str(i)+'LT3.ASP', 'r') as file:
        for line_number, line in enumerate(file, start=1):
            if line_number >= 7 and line_number < 308:
                line =line.strip()
                try:
                # print(lin)
                    perL = float(line.strip())
             #       print(line_number, line)
                    Tper_L[i-1].append(perL)


                except ValueError:
                    print(f"skipping invalid line: {line}")

            #Transverse

    with open('B'+str(i)+'TT3.ASP', 'r') as file:
        for line_number, line in enumerate(file, start=1):
            if line_number >= 7 and line_number < 308:
                line =line.strip()
                try:
                # print(lin)
                    perT = float(line.strip())
             #       print(line_number, line)
                    Tper_T[i-1].append(perT)


                except ValueError:
                    print(f"skipping invalid line: {line}")
                    
                    
    plt.figure(i)
    plt.plot(lam, Tper_L[i-1], label = 'Longitudinal')  
    plt.plot(lam, Tper_T[i-1], label = 'Transverse')  
    plt.xlabel('Wavelength')
    plt.ylabel('Transmission percentage')
    plt.title('Transmittance BGO(' + str(i)+')' )
    plt.grid(True)
    plt.legend()
#print(Tper)

plt.show()
plt.ion()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Generate some example data
x = np.linspace(0, 10, 1000)
y = np.sin(x) + 0.3 * np.random.randn(1000)  # Add some noise to the sine wave

# Find peaks in the data
peaks, _ = find_peaks(y)

# Plot the data and highlight the peaks
plt.plot(x, y, label='Signal')
plt.plot(x[peaks], y[peaks], 'ro', label='Peaks')  # Red dots for peaks
plt.title("Peak Detection Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
